```

#on definit une fonction qui renvoie les id des artists
import requests
def get_id(artist):
  url=f"https://api.happi.dev/v1/music?q={artist}&limit=&type=:type&lyrics=0"
  api_key="2140cctW6tkwuPj4oVRnUexS1gqK8t1d30B0XVsNkFIqSYvgnDJjwzXI"
  r=requests.get(url, params={"apikey":api_key})
  data=r.json()
  return data["result"][0]["id_artist"]
```

On réitère le même procédé que l'on a fait pour récupérer les id:

```
# https://api.happi.dev/v1/music/artists/468/smart-playlist?apikey=2140cctW6tkwuPj4oVRnUexS1gqK8t1d30B0XVsNkFIqSYvgnDJjwzXI
#playlist de Daft Punk
```

```
def get_playlist(artist_id):
  artist_track=[]
  url=f"https://api.happi.dev/v1/music/artists/{artist_id}/smart-playlist?"
  api_key="2140cctW6tkwuPj4oVRnUexS1gqK8t1d30B0XVsNkFIqSYvgnDJjwzXI"
  r=requests.get(url, params={"apikey":api_key})
  data=r.json()
  artist_track.append(data["result"][0]["artist"])
  #for i in range(len(data["result"])):
    #artist_track.append(data["result"][i]["track"])
  artist_track.append(data["result"][1]["track"])
  return artist_track
```

# Geocoding
Le géocoding permet de géolocaliser à travers une carte vos contacts à partir d'une adresse.

```
import requests

link = "https://api-adresse.data.gouv.fr/search/?q=+44+rue+alphonse+penaud+75020"
r=requests.get(link).json()
r
```

```
def API_adresse(adresse):
  link=f"https://api-adresse.data.gouv.fr/search/?q={adresse}"
  r=requests.get(link).json()
  return r["features"][0]["geometry"]["coordinates"][::-1]
```

```
API_adresse("44 rue alphonse penaud, 75020")
```

On récupère bien les points de coordonnées des adresses correspondantes.        
 Essayons à présents de les afficher sur une carte.  
Pour ce faire, on va utiliser la librairie "Folium".

```
import folium 
m=folium.Map(location=[48, 2], zoom_start=14)
m
```

```
folium.Marker(location=[48.868851, 2.40604], popup="Bédi's Pizzaria").add_to(m)
m
```

```
import pandas as pd

cinema_creuse = pd.DataFrame([["Cinéma Claude Miller","Place Mail, 23400 Bourganeuf"],
                              ["Cinéma Le Sénéchal", "1 Rue du Sénéchal, 23000 Guéret"],
                              ["Cinéma Eden", "4 Place Saint-Jacques, 23300 La Souterraine"],
                              ["Cinéma Alpha", "19 Rue de Rentière, 23110 Évaux-les-Bains"],
                              ["Cinéma le Colbert","50 Grande Rue, 23200 Aubusson"],
                              ["Cinéma Le Marchois", "17 Rue des Deux Frères, 23100 La Courtine"]],
                             columns = ["nom", "adresse"]
                             )
cinema_creuse
```

```
cinema_creuse["coordonnees"]=cinema_creuse["adresse"].apply(API_adresse)
```

```
for i in range(0, len(cinema_creuse)):
  folium.Marker(
      location=cinema_creuse.loc[i, "coordonnees"], popup=cinema_creuse.loc[i, "nom"], icon=folium.Icon(color="red" ,icon="pushpin")).add_to(m)
m
```
